# matplotlib

##### Gráficos de Linha

```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
amount = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
# cria gráfico de linha
plt.plot(years, amount, color = 'blue', marker = 'o', linestyle = 'solid')

# titulo
plt.title("Amount")

# label do eixo y
plt.ylabel("Bilhoes $")

plt.show()
```

##### Gráficos de Barra

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# adiciona-se 0.5 às coordenadas à esquerda para que cada barra seja centralizada
xs = [i + 0.5 for i, _ in enumerate(movies)]

# barras com as coordenadas x à esquerda [xs], alturas [num_oscars]
plt.bar(xs, num_oscars)

plt.ylabel("# de premiações")
plt.title("Filmes")

# nomeia eixo x com nomes de filmes na barra central
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

plt.show()
```
